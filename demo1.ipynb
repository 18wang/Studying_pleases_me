## 1.数值运算

```python
#数值运算,多项式方程的求解
var('x,a,b,c,d')  #定义变量

#三次方程求根
f(x)=a*x^3+b*x^2+c*x+d
show(f(x)==0)

solutions=solve(f(x)==0,x, solution_dict=true)
for s in solutions:
    show(s)
```
